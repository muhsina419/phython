2.Check whether a number is palindrome or not

```python
#integer
a=int(input("Enter the number : "))
m=a
sum=0
while(m!=0):
  r=m%10
  m=int(m/10)
  sum=sum*10+r
if(sum==a):
  print(a," is a palindrome")
else:
  print(a," is not a palindrome")
```
