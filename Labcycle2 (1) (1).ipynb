2.Write a program to read a string containing numbers separated by a space and convert it as a list of integers.Perform the following operations on it.


1.   Rotate elements in a list by 'k' position to the right
2.   Convert the list into  a tuple using list comprehension
3.   Remove all duplicates from the tuple and convert them into a list again.
4.  Create another list by putting the results of evaluation of the function
 f(x)=x^2-x with each element in the final list
5.   After sorting them individually, merge the two lists to create a single sorted list






```python
def rotate_list(nums, k):
    k = k % len(nums)
    return nums[-k:] + nums[:-k]

def remove_duplicates(t):
    return list(set(t))

def evaluate_function(nums):
    return [x**2 - x for x in nums]

input_string = input("Enter a string of numbers separated by a space: ")
k = int(input("Enter the number of positions to rotate: "))

numbers = list(map(int, input_string.split()))
print("Original List:", numbers)

rotated_numbers = rotate_list(numbers, k)
print("Rotated List:", rotated_numbers)

numbers_tuple = tuple([num for num in rotated_numbers])
print("Tuple:", numbers_tuple)

unique_numbers = remove_duplicates(numbers_tuple)
print("Unique Numbers List:", unique_numbers)

evaluated_numbers = evaluate_function(unique_numbers)
print("Evaluated Numbers List:", evaluated_numbers)

unique_numbers.sort()
evaluated_numbers.sort()

merged_sorted_list = sorted(unique_numbers + evaluated_numbers)
print("Merged Sorted List:", merged_sorted_list)
```

4. Write a program to create a class Box with data members length,
breadth, height, area, and volume. Provider constructor that enables
initialization with one parameter (for cube), two parameters (for
square prism) three parameters (rectangular prism). Also, provide
functions to calculate area and volume.
Create a list of N boxes with random measurements and print the
details of the box with maximum volume: area ratio.


```python
class box:
    def __init__(self,length,breadth=None,height=None):
        #rectangular prism
        self.Length=length
        self.Breadth=breadth
        self.Height=height

        #cube
        if breadth is None and height is None:
            self.Breadth=length
            self.Height=length
        elif height is None:
            self.Height=length

    def Area(self):
        area=2*(self.Length*self.Breadth+self.Breadth*self.Height+self.Length*self.Height)
        return area

    def volume(self):
        volu=self.Length*self.Breadth*self.Height
        return volu

cube=box(6)
sq_prism=box(3,8)
rect_prism=box(6,3,8)
print("AREA OF CUBE:", cube.Area())
print("VOLUME OF CUBE:",cube.volume())

print("AREA OF SQUARE PRISM:", sq_prism.Area())
print("VOLUME OF SQUARE PRISM:",sq_prism.volume())
print("AREA OF RECTANGULAR PRISM:",rect_prism.Area())
print("VOLUME OF RECTANGULAR PRISM:",rect_prism.volume())
```
